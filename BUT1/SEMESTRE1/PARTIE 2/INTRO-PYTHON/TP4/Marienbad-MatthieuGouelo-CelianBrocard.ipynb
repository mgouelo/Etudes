```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri Jan 21 10:32:50 2022

@author: druet
"""

import random
import time


################### le message de bienvenue et d'explication ##################

def welcome_message():
    print("""
----------------------------------------------------------------------------
                        Règles du jeu de Nim
----------------------------------------------------------------------------
Règles : 1 - L'utilisateur et l'ordinateur jouent à tour de rôle.
         2 - L'ordinateur commence.
         3 - Seulement 1 ou 2 bâtons peuvent être retirés à chaque coup.
         4 - Celui qui tire le dernier bâton a gagné.
         
----------------------------------------------------------------------------
                           Apprentissage
----------------------------------------------------------------------------                        
Suivant qu'il gagne ou perd, l'ordinateur reçoit une récompense ou une punition.
----------------------------------------------------------------------------
""")

#### Nous faisons commencer l'ordinateur parce que le premier joueur a une 
#### stratégie gagnante. Si le joueur commence et qu'il joue bien, la machine 
#### perdra quoiqu'elle fasse et n'apprendra rien. 

###############################################################################


################## définition et impression de la position de jeu #############

def printboard(n):
    board=[]
    for _ in range(n):
        board.append("/")
    print("\n----------------------------------------------------------------------------")
    print("      ",*board, sep="   ")
    print("----------------------------------------------------------------------------\n")
    print("Il reste " + str(n) + " allumettes.")
    
###############################################################################


################### initialisation de la machine ##############################

nombre_allumettes=8 ### vous pouvez changer le nombre d'allumettes de départ
board = [] ### le plateau de jeu
boulesjaunes = [] ### correspond au nombre de boules jaunes dans la case
boulesrouges = [] ### correspond au nombre de boules rouges dans la case
tirage = [] ### correspondra au tirage dans une partie
for _ in range(nombre_allumettes):
    boulesjaunes.append(2)
    boulesrouges.append(2)
    tirage.append(0)
#### Attention Python commence à 0 #####
boulesrouges[0]=0 #### il ne faut pas mettre de boules rouges dans la case 0, coup interdit


###############################################################################


#################### programme principal : jeu + renforcement #################
welcome_message()
uneautrepartie=True
compteur_partie=0
time.sleep(1)
print("La probabilité de gain de l'ordinateur à la première partie si le joueur joue optimalement est de 12,5 %")
time.sleep(1)
while uneautrepartie :
    player = "CMP"
    allumettes=nombre_allumettes
    tirage=[0,0,0,0,0,0,0,0]
    while allumettes>0:
        printboard(allumettes)  ### imprime la position de jeu
        time.sleep(1)
        if player=='CMP': ### c'est à l'ordinateur de jouer
            print("\nL'ordinateur choisit de retirer...")
            time.sleep(1)
            somme=boulesjaunes[allumettes-1]+boulesrouges[allumettes-1]
            boulehasard=random.randint(1,somme)##permet de tirer jaune ou rouge
            if boulehasard <= boulesjaunes[allumettes-1]: ##la machine a tiré jaune, i.e. elle enlève une allumette
                tirage[allumettes-1]=1
                allumettes=allumettes-1
                print("1 allumette.")
            else:
                tirage[allumettes-1]=2
                allumettes=allumettes-2
                print("2 allumettes.")
            if allumettes==0:
                winner='CMP'
            else:
                player='USER'  
        else: ### c'est au joueur de jouer
#### on demande au joueur ce qu'il souhaite jouer avec vérification que c'est un coup légal
            coup_joueur=0
            print("\n--À vous de jouer !--")
            while coup_joueur not in range(1, 3) or coup_joueur>allumettes:
                try:
                    coup_joueur = int(input("\nQuel est votre choix ?"))
                    if coup_joueur == 0:
                        print("\nVous devez enlever au moins une allumette !")
                    elif coup_joueur not in range(1, 3) or coup_joueur>allumettes:
                        print("\nVous ne pouvez pas enlever autant d'allumettes !")
                        coup_joueur = int(input("\nQuel est votre choix ?"))
                except Exception as e:
                    print("\nCela ne semble pas une réponse valide.\nError: " + str(e) + "\nRecommencez !")
##### fin du choix du joueur
            allumettes=allumettes-coup_joueur
            if allumettes==0:
                winner='USER'
            else:
                player='CMP' 
    compteur_partie+=1
##### fin de la partie ######
##### annonce des résultats #####
    if winner=='CMP':
        print("\n----------------------------------------------------------------------------")
        print("L'ordinateur a gagné, nous allons le récompenser.")
        print("----------------------------------------------------------------------------\n")
    else:
        print("\n----------------------------------------------------------------------------")
        print("Bravo ! Vous avez gagné ! Nous allons punir l'ordinateur.")
        print("----------------------------------------------------------------------------\n")
##### Apprentissage : récompense ou punition de l'ordinateur#####        
    if winner=='CMP': ###récompense
        for i in range(nombre_allumettes):
            if tirage[i]==1:
                boulesjaunes[i]=boulesjaunes[i]+1
            if tirage[i]==2:
                boulesrouges[i]=boulesrouges[i]+1
    else: ###punition
        for i in range(nombre_allumettes):
            if tirage[i]==1:
                boulesjaunes[i]=boulesjaunes[i]-1
            if tirage[i]==2:
                boulesrouges[i]=boulesrouges[i]-1
####### fin de la récompense ou de la punition #################
####### réinitialisation des verres vides ######################
    for i in range(nombre_allumettes):
        if (boulesjaunes[i]==0) and (boulesrouges[i]==0):
            boulesjaunes[i]=2
            boulesrouges[i]=2
######## impression de l'état des verres #######################
    time.sleep(1)
    for i in range(nombre_allumettes):
        print("Dans le verre " + str(i+1) +", il y a " + str(boulesjaunes[i]) + " boules jaunes et " + str(boulesrouges[i]) + " boules rouges.")
################################################################
############ calcul de la probabilité de gagner ################
    time.sleep(1)
    if (boulesjaunes[3]/(boulesjaunes[3]+boulesrouges[3]))>(boulesrouges[4]/(boulesjaunes[4]+boulesrouges[4])):
        proba=(boulesrouges[7]/(boulesjaunes[7]+boulesrouges[7]))*(boulesrouges[4]/(boulesjaunes[4]+boulesrouges[4]))*(boulesrouges[1]/(boulesjaunes[1]+boulesrouges[1]))
    else:
        proba=(boulesrouges[7]/(boulesjaunes[7]+boulesrouges[7]))*(boulesjaunes[3]/(boulesjaunes[3]+boulesrouges[3]))*(boulesrouges[1]/(boulesjaunes[1]+boulesrouges[1]))
    print("\n----------------------------------------------------------------------------")
    print("Vous avez joué "+ str(compteur_partie) +" parties.")
    print("----------------------------------------------------------------------------\n")
    print("\n----------------------------------------------------------------------------")
    print("La probabilité de gain de l'ordinateur à la prochaine partie si le joueur joue optimalement (en connaissant l'état des verres :-)) est de " + str(round(proba*100,2)) + "%")
    print("----------------------------------------------------------------------------\n")
################################################################
############ On continue ? ################    
    test=True
    while test:
        another_go = input("\nVoulez-vous rejouer ?[O/N]: ")
        if another_go in ("o","O"):
            uneautrepartie=True
            test=False
        elif another_go in ("n","N"):
            uneautrepartie=False
            test=False
        else:
            print("\nChoix invalide. Recommencez !")    
############################################
```

## **JEU DE NIM IA**

Joueur MACHINE : stratégie pseudo aléatoire \
Joueur IA : Joue une première fois aléatoirement puis adapte son jeu en fonction des parties précédentes \
\
la variable _logPlay_ permet d'enregistrer l'historique des coups des anciennes parties selon le nombre d'allumettes restante. Si une partie est perdue, tous les coups joués vont être décrémentés de 1 dans _logplay_ ; C'est la punition \
A l'inverse si une partie est gagnée, tous les coups joués vont être incrémentés de 1 dans _logplay_. \
Au bout d'un grand nombre de parties, on devrai remarquer que certain coups s'avèrent plus interessants que d'autre selon le nombre d'allumettes restant. C'est ainsi qu'on verra se dessiner la stratégie gagnante \
\
A chaque début de partie, on crée une liste _historique_ construite de la même manière que logPlay. Cette liste enregistre les coups jouer dans la partie actuelle uniquement en remplacant les 0 initialisés par défaut par des 1.

- Si la partie est gagnée, on ajoute la liste des coups joués _historique_ à _logplay_ qui enregistre les coups joués sur toutes les parties
- Si la partie est perdue, on soustrait la liste des coups joués _historique_ à logPlay.\
(pour créer _logPlay_ et _historique_ on utilisera la même fonction soit **creerHistorique** car leur construction est identique)


```python
import random

def init_niammi():
    players = ["MACHINE", "IA"]
    currentPlayer = random.choice(players)
    return (players, currentPlayer)



def gagne(l):
    gagner = True
    i = 0
    while i < len(l) and gagner == True:
        if l[i] != 0: #s'il reste encore des allumettes
            gagner = False
        else:
            i += 1
    return gagner



def switch_player(players, currentPlayer):
    if currentPlayer == players[0]:
        currentPlayer = players[1]
    else:
        currentPlayer = players[0]
    return currentPlayer



def pseudo_aleatoire(l):
    '''
    Stratégie utilisée par la machine jouant contre l'IA. Cette fonction permet à la machine de jouer 
    aléatoirement sauf si un coup plus intelligent permettrai de bloquer l'adversaire ou de gagner directement
    '''
    # Vérifier s'il est possible de gagner directement
    somme = sum(l)
    
    for ligne in range(len(l)):
        if somme <= 3: #Il reste plus que 3 ou moins d'allumettes en jeu. Il est peut être possible de gagner directement...
            if l[ligne] == somme:  # Si toutes les allumettes restantes sont sur la même ligne on les retire et c'est gagné !
                l[ligne] -= l[ligne]
                return l
            elif l[ligne] == 2 and somme == 3:# 1ère ligne à 1 allumettes et 2e ligne à 2 allumettes alors ...
                l[ligne] -= 1
                return l
        else:
            # Calculer un coup stratégique : garder un multiple de 4 (si possible)
            for nbAllu in range(1, 4):
                if l[ligne] > 4 and (l[ligne] - nbAllu) % 4 == 0 and l[ligne] - nbAllu >= 0:
                    l[ligne] -= nbAllu
                    return l

    # Sinon, jouer de manière aléatoire
    ligne = random.randint(0, len(l) - 1)
    while l[ligne] == 0:  # Trouver une ligne avec des allumettes
        ligne = random.randint(0, len(l) - 1)

    # Sinon, jouer un coup aléatoire valide
    nbAllu = random.randint(1, 3)
    while (l[ligne] - nbAllu) < 0:
        nbAllu = random.randint(1, 3)

    l[ligne] -= nbAllu
    return l



def jouer_smart(l, historique, logPlay):
    '''
    Stratégie utilisée par l'IA. La stratégie consiste à comparer les coefficient dans la sous liste correspondant au bon nombre 
    d'allumettes restant. Si un coefficient est plus grand que les autres alors son indice dans la sous liste + 1 donne le nombre
    d'allumette à retirer. Ce coup est enregistrer dans la variable historique
    '''
    somme = sum(l)
    if logPlay[somme-1][0] == logPlay[somme-1][1] == logPlay[somme-1][2] :
        nbAllu = random.randint(1, 3)

    else :
        maxi = max(logPlay[somme-1])
        nbAllu = logPlay[somme-1].index(maxi) + 1 #On prends l'indice du max + 1 pour obtenir le nombre d'allumettes à retirer

    # Maintenant qu'on connait le nombre d'allumettes à retirer, on cherche une ligne qui contient un nombre suffisant d'allumettes pour jouer ce coup
    while nbAllu > 0:
        for ligne in range(len(l) - 1, -1, -1):  # Parcours des lignes de la fin au début
            if l[ligne] >= nbAllu:  # Ligne valide
                l[ligne] -= nbAllu
                historique[somme - 1][nbAllu - 1] += 1  # Enregistrement dans l'historique
                return l, historique

        # Si aucune ligne est trouvée, on réduit le nombre d'allumettes retiré
        nbAllu -= 1



def creerHistorique(l):
    '''
    LogPlay construction :
    - Les indices dans la liste correspondent au nombre d'allumettes restante -1
    - Les indices des listes imbriquées correspondent au nombre d'allumettes retirer dans les précédentes parties dans la même configuration - 1
    - Exemple : s'il reste 14 allumettes en jeu et que l'IA retire 2 allumettes et que finalement la partie est gagnée alors on incrémente de 1
    logPlay[13][1] car ce coup et tous les autres joué au cours de la partie on mené vers la victoire.
    LogPlay enregistre
    '''
    historique = []
    somme = sum(l)

    for i in range(somme):
        historique.append([0, 0, 0])

    # On met toutes les valeur interdites à -∞ (exemple : retirer 2 allumettes alors qu'il en reste qu'une)
    # Dans la liste historique l'ia va privilègier le coups qui possède le plus grand nombre associé. 
    # En mettant les valeurs interdites à -∞, on est sur que l'ia ne choisira jamais ces coups
    historique[0][1] = historique[0][2] = historique[1][2] = -float('inf') 

    return historique # Les indices dans la liste correspondent au nombre d'allumettes restante -1



def bornes(logPlay):
    '''
    Permet de limiter les coefficients de logplay sur l'intervalle [-10;10]. Avoir une étendu faible permet d'adapter plus rapidemment 
    la stratégie que si les coefficiens pouvait atteindre plusieurs milliers
    '''

    for i in range(len(logPlay)):
        for j in range(len(logPlay[i])):
            if (i == 0 and j == 1) or (i == 0 and j == 2) or (i == 1 and j == 2): #les indices des valeurs interdites on les laisse telles quelles (On veut éviter que les valeurs interdites soient choisies par l'ia)
                pass
            else:
                if logPlay[i][j] < -10:
                    logPlay[i][j] = -10
                elif logPlay[i][j] > 10:
                    logPlay[i][j] = 10
    return logPlay
```

```python
def niammi(n, logPlay):
    l = [1, 3, 5, 7]
    ia = 0 # nombre de partie gagner par l'ia
    machine = 0 # nombre de partie gagner par la machine
    for i in range(n): # avec n le nombre de partie qu'on veut simuler
        running = True
        l = [1, 3, 5, 7]
        historique = creerHistorique(l)
        players, currentPlayer = init_niammi()
        while running:
            if currentPlayer == "IA":
                l, historique = jouer_smart(l, historique, logPlay) #Si l'IA joue on utilise jouer_smart()
            else:
                l = pseudo_aleatoire(l) # sinnon c'est la machine qui joue

            gagner = gagne(l) # On vérifie si la partie est terminé
            if gagner:
                running = False
                if currentPlayer == "IA":
                    recompense = True #Permet de déterminé si on ajoute l'historique à logplay pour insiter l'IA à rejouer ces coups ou au contraire on soustrait pour l'en dissuader
                    ia += 1
                else:
                    recompense = False
                    machine += 1
            else:
                currentPlayer = switch_player(players, currentPlayer)

        for i in range(len(logPlay)):
            for j in range(len(logPlay[i])):
                if recompense:
                    logPlay[i][j] += historique[i][j] # On additionne car les coups joués on mené à la victoire
                else:
                    logPlay[i][j] -= historique[i][j] * 2 # On soustrait car les coups joués on mené à la défaite et on multiplie par 2 pour une punition plus strict, exigeante pour obtenir un résultat plus précis

        logPlay = bornes(logPlay)

    print(f"IA : {ia} | MACHINE : {machine}")
    print("Pourcentage de victoire : ", ia*100/(ia+machine), "%")
    return logPlay
```
